```python
data = coredata[['country','sex','age_years','occupation','recovered_y_n']].drop("occupation",axis=1)
data.head()
```

```python
sns.heatmap(data.isnull())
```

```python
data1 =data.dropna()
data1.hist()
```

```python
sns.heatmap(data1.isnull())
```

```python
data1['gender'] = np.where(data1['sex']=="Male",1,0)
data1['nonsurvival'] = np.where(data1['recovered_y_n']=="y",0,1)
```

```python
data1
```

```python
Y = data1['nonsurvival']
X = data1[['age_years','gender']]
```

```python
import statsmodels.api as sm
X = sm.add_constant(X)
X
```

```python
model_1 = sm.Logit(Y,X).fit()
```

```python
model_1.summary()
```

```python
pred= model_1.predict()
pred
sns.scatterplot(x=data1['nonsurvival'],y=pred)
```

```python
from sklearn.metrics import classification_report,balanced_accuracy_score
```

```python
data1['predcoded'] = np.where(pred > 0.23,1,0)
data1['pred'] = pred
print(balanced_accuracy_score(data1['nonsurvival'],data1['predcoded']))
print(classification_report(data1['nonsurvival'],data1['predcoded']))
```

```python
sns.jointplot(x="pred",y='nonsurvival',data=data1,kind="hex")
```

```python
coef=model_1.params
coef
```
